```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from sklearn.metrics import mean_squared_error
from math import sqrt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose

from pandas import Series
from matplotlib import pyplot
```

```python
df = pd.read_csv("temperatures.csv")
```

```python
# Additive decomposition

result = seasonal_decompose(df['Temp'], model='additive', freq=1)
result.plot()
pyplot.show()
```

```python
train = df[:int(0.7*(len(df)))]
test = df[int(0.7*(len(df))):]
```

```python
#Plotting data
train.plot(figsize=(15,8), title= 'Temperatures', fontsize=14)
test.plot(figsize=(15,8), title= 'Temperatures', fontsize=14)
plt.show()
```

```python
# Single exponential smoothing

y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['Temp'])).fit(smoothing_level=0.6,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot(train['Temp'], label='Train')
plt.plot(test['Temp'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Temp, y_hat_avg.SES))
print(rms)
```

```python
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['Temp']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['Temp'], label='Train')
plt.plot(test['Temp'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Temp, y_hat_avg.Holt_Winter))
print(rms)
```
